```python
hours = []
minutes = []
seconds = []
for i in range(df.shape[0]):
    try:
        time = parser.parse(df['Launch Time (UTC)'][i])
        hours.append(time.hour)
        minutes.append(time.minute)
        seconds.append(time.second)
    except:
        hours.append(0)
        minutes.append(0)
        seconds.append(0)
```

```python
df['hours'] = hours
df['minutes'] = minutes
df['seconds'] = seconds
```

```python
months = []
years = []
days = []
for i in range(df.shape[0]):
    try:
        date = parser.parse(df['Date'][i])
        months.append(date.month)
        years.append(date.year)
        days.append(date.day)
    except:
        months.append(0)
        years.append(0)
        days.append(0)
```

```python
df['month'] = months
df['day'] = days
df['year'] = years
```

```python
months = []
years = []
days = []
for i in range(df.shape[0]):
    try:
        date = parser.parse(df['Date Entered/Updated'][i])
        months.append(date.month)
        years.append(date.year)
        days.append(date.day)
    except:
        months.append(0)
        years.append(0)
        days.append(0)
```

```python
df['month Entered/Updated'] = months
df['day Entered/Updated'] = days
df['year Entered/Updated'] = years
```

```python
df.drop(['Launch Time (UTC)','Date','Date Entered/Updated'],axis=1, inplace=True)
```

```python
df['Facility Latitude'] = df['Facility Latitude'].replace('Unknown' , 0)
```

```python
df['Facility Longitude'] = df['Facility Longitude'].replace('Unknown' , 0)
```
