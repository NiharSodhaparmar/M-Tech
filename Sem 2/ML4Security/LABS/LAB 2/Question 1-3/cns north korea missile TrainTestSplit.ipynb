# Import datetime module

```python
from datetime import datetime
from dateutil import parser
```

# Process Launch Time (UTC) Column

```python
hours = []
minutes = []
seconds = []
count = 0
for i in range(df.shape[0]):
    try:
        time = parser.parse(df['Launch Time (UTC)'][i])
        hours.append(time.hour)
        minutes.append(time.minute)
        seconds.append(time.second)
    except:
        count = count + 1
        hours.append(0)
        minutes.append(0)
        seconds.append(0)
print(count)
```

```python
df['hours'] = hours
df['minutes'] = minutes
df['seconds'] = seconds
```

```python
df.drop(['Launch Time (UTC)'],axis=1, inplace=True)
```

# Process DATE Column

```python
months = []
years = []
days = []
count = 0
for i in range(df.shape[0]):
    try:
        date = parser.parse(df['Date'][i])
        months.append(date.month)
        years.append(date.year)
        days.append(date.day)
    except:
        count = count + 1
        months.append(0)
        years.append(0)
        days.append(0)
print(count)
```

```python
df['month'] = months
df['day'] = days
df['year'] = years
```

```python
df.drop(['Date'],axis=1, inplace=True)
```

# Process Facility Latitude and Facility Longitude column

```python
col = ['Facility Latitude' ,'Facility Longitude' ]

for i in col:
    df[i] = df[i].replace('Unknown' , 0)
    df[i] = df[i].astype(float)
```

# Replace all unknown value with mode or mean

```python
mean_replace_col = ['Facility Latitude','Facility Longitude','Apogee','Distance Travelled']

for i in mean_replace_col:
    df[i] = df[i].replace(0 , df[i].mean())
```

```python
mode_replace_col = ['Missile Name' , 'Landing Location','Missile Type','Facility Location','Facility Name','Missile Type' , 'Test Outcome']

for i in mode_replace_col:
    df[i] = df[i].replace('Unknown' , df[i].mode()[0])
```
