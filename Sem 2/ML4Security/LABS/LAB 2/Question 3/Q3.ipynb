# Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import IsolationForest
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

# LogisticRegression

```python
logisticRegression = LogisticRegression()
logisticRegression.fit(X_train, y_train)
```

# DecisionTreeClassifier

```python
decisionTree = DecisionTreeClassifier()
decisionTree.fit(X_train, y_train)
```

# IsolationForest

```python
isolationForest = IsolationForest()
isolationForest.fit(X_train, y_train)
```

# SVC

```python
svm = SVC()
svm.fit(X_train, y_train)
```

# KNeighborsClassifier

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

# GaussianNB

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
```

```python
ypred_logistic = logisticRegression.predict(X_validate)
ypred_decision = decisionTree.predict(X_validate)
ypred_isolation = isolationForest.predict(X_validate)
ypred_svc = svm.predict(X_validate)
ypred_knn = knn.predict(X_validate)
ypred_gaussian = gaussian.predict(X_validate)
```

```python
from sklearn.metrics import accuracy_score
print("Logistic Regression : ", accuracy_score(y_validate, ypred_logistic))
print("Decision Tree : ", accuracy_score(y_validate, ypred_decision))
print("Isolation Forest : " , accuracy_score(y_validate, ypred_isolation))
print("svm : " , accuracy_score(y_validate, ypred_isolation))
print("knn : " , accuracy_score(y_validate, ypred_knn))
print("gaussian : " , accuracy_score(y_validate, ypred_gaussian))
```
