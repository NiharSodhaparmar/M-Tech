```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("citrus.csv")
df.head()
```

```python
df.info()
```

```python
df['name'].unique()
```

```python
le  = LabelEncoder()
```

```python
df['name'] = le.fit_transform(df['name'])
```

```python
df.head()
```

```python
X = df.drop(['name'], axis=1)
X = np.asarray(X).astype(np.float32)
y =df['name']
y = np.asarray(y).astype(np.float32)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# optimizer = Adem, loss = BinaryCrossentropy

#### activation = (relu , sigmoid)

```python
inputs = keras.Input(shape=(5,), name="input")
x1 = layers.Dense(16, activation="relu")(inputs)
x2 = layers.Dense(8, activation="relu")(x1)
x3 = layers.Dense(4, activation="relu")(x2)
outputs = layers.Dense(1, activation='sigmoid', name="predictions")(x3)
model = keras.Model(inputs=inputs, outputs=outputs)
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.001)
loss_fn = keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
```

```python
epochs = 100
batch_size = 64
results1_1 = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2, verbose=1)
```

```python
_ , accuracy1_Adam_BinaryCrossentropy=model.evaluate(X_test, y_test, verbose=1)
accuracy1_Adam_BinaryCrossentropy
```
