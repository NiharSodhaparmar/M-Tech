```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
citrus_data = pd.read_csv("./citrus.csv")
```

```python
columns = ["name", "diameter", "weight", "red", "green", "blue"]
```

```python
citrus_data.shape
```

```python
citrus_data
```

## Pie chart

```python
labels = ["Orange", "Grapefruit"]
sizes = list(citrus_data["name"].value_counts())
plt.pie(sizes, labels=labels)
plt.axis("equal")
plt.legend()
```

## Barchart

```python
for i in columns:
    if i!= "name":
        plt.figure()
        sns.barplot(citrus_data, x=citrus_data["name"], y=citrus_data[i])
        plt.title("Barplot")
```

## Countplot

```python
sns.countplot(x="name", data=citrus_data)
plt.title("Countplot for name")
```

## Scatterplot

```python
for i in columns:
    for j in columns:    
        if i == "name" or j == "name":
            continue
        else:
            plt.scatter(citrus_data[i], citrus_data[j], color="green")
            plt.title("Scatter Plot for bivariate analysis")
            plt.xlabel(i)
            plt.ylabel(j)
            plt.show()
```

## Histogram

```python
for i in columns:
    if i!= "name":        
        plt.figure()
        plt.hist(citrus_data[i], color="green", bins = 25)
        plt.title("Histogram")
        plt.xlabel(i)
        plt.ylabel("Frequency")
```

## Boxplot

```python
for i in columns:
    if i != "name":
        plt.figure()
        sns.boxplot(x=citrus_data[i], data=citrus_data)
        plt.title("Boxplot showing distribution for " + i)
```

## Heatmap

```python
sns.heatmap(citrus_data.corr(), cmap='coolwarm', annot=True)
```

## Displot

```python
for i in columns:
    if i != "name":
        sns.displot(citrus_data[i], bins=25, kde=True)
```

## Pairplot

```python
sns.pairplot(citrus_data, hue="name")
```
