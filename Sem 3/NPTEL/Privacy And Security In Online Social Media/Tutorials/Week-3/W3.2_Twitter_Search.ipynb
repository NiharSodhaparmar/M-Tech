# Search API

```python
import tweepy
from keys import *
from pprint import pprint
```

```python
#Consumer Key (API Key), Consumer Secret (API Secret)
auth = tweepy.OAuthHandler(Consumer_Key, 
                           Consumer_Secret)
# Access Token, Access Token Secret
auth.set_access_token(Access_Token, 
                      Access_Token_Secret)

api = tweepy.API(auth)
if (not api):
    print("Authentication failed :(")
else:
    print("Authentication successfull!!! :D")
```

### Search API

#### Query

```python
query = 'delhi winter'
```

#### Query the endpoint

```python
search_results = api.search_tweets(q=query)
```

#### Print the result

```python
type(search_results)
```

## Filters

```python
query = 'delhi winter' 
hi_lang = 'hi' 
popular_results = 'popular' 
extended_mode = 'extended'
```

```python
search_results = api.search_tweets(q=query, lang=hi_lang, result_type=popular_results, 
                            tweet_mode=extended_mode)
```

```python
for tweet in search_results:
    pprint(tweet.full_text)
    print("-"*100)
```

### Save the tweets

```python
import json
```

#### Iterate through search results and save the tweet

```python
for idx,tweet in enumerate(search_results):
    with open(f"tweet_dump/tweet_{idx}.json",'w') as fp:
        json.dump(tweet._json,fp,indent=4)
```
