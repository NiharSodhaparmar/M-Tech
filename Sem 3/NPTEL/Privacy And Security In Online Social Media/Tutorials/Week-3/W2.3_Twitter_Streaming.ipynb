```python
counter=100
#override tweepy.StreamListener to add logic to on_status
class MyStreamListener(tweepy.Stream):
        
    def on_status(self, status):
        global counter
        print(status.text)
        print(f"Saving to file tweet_dump/tweet_{counter}.json")
        with open(f"tweet_dump/tweet_{counter}.json",'w') as fp:
            json.dump(status._json,fp,indent=4)
        counter+=1
        print("-"*100)
```

```python
myStream = MyStreamListener(consumer_key=Consumer_Key,consumer_secret=Consumer_Secret,access_token=Access_Token,access_token_secret=Access_Token_Secret)
```

```python
myStream.filter(track=['iphone'])
```
