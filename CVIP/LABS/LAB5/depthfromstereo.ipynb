##### Step-1 Load necessary packages

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Step-2 â€” Load the left and right images in grayscale

```python
imgLeft = cv2.imread('Left Image.png', 0)
imgRight = cv2.imread('Right Image.png', 0)
```

##### Step-3 Display the two input images

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(imgLeft, 'gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(imgRight, 'gray')
plt.axis('off')
```


##### Step-4 Define a function to calculate and show disparity

```python
def ShowDisparity(bSize=5):
    # Initialize the stereo block matching object 
    stereo = cv2.StereoBM_create(numDisparities=32, blockSize=bSize)

    # Compute the disparity image
    disparity = stereo.compute(imgLeft, imgRight)

    # Normalize the image for representation
    min = disparity.min()
    max = disparity.max()
    disparity = np.uint8(255 * (disparity - min) / (max - min))
    
    # Plot the result
    return disparity
```

##### Step-5 - Perform depth estimation

```python
result = ShowDisparity(bSize=5)
plt.imshow(result, 'gray')
plt.axis('off')
plt.show()
```

```python
result = ShowDisparity(bSize=25)
plt.imshow(result, 'gray')
plt.axis('off')
plt.show()
```
